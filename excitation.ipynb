## **Calculate the excitation potencial**

Methods suggested:

*   Method 1: Difference between peaks. Get the mean and standard deviation. Is there any difference from one Vr value to the outher?

*   Method 2: Peak position obtained using squared fit and the excitation potencial obtaind by linear fit. (Choose one temperature).

*   Method 3: Fourier Transform (FFT). (Choose one temperature).
   
## **Calculate the excitation potencial**

For each method it is calculated the contact potencial. This potencial is calculated using the value found for the excitation potencial and the difference between this value and the position of the first peak.

## **Calculate the Hg transition wavelenght**

For each method it is calculated the Hg transition wavelenght from the excitation potencial and compared with the theoritical value 253.7 nm:

λ = ch/E, whre E is the excitation potencial in eV.

```python
def par(x,a,b,c):
    return a*x**2+b*x+c



def fit_plot(x, y):
    popt, pcov = curve_fit(par,x,y)
    a,b,c = popt

    correlation_matrix = np.corrcoef(x, y)
    correlation_xy = correlation_matrix[0,1]
    r_squared = correlation_xy**2
    error = np.sqrt(np.diagonal(pcov)) #incerteza dos parametros
    x_std = np.std(x)
    V_x = -popt[1]/(2*popt[0]) # x coordinate vertex
    V_y = (-popt[1]**2-4*popt[0]*popt[2])/(4*popt[0]) # y coordinate vertex
    peak = (V_x,V_y)

    plt.rcParams["figure.figsize"] = [10, 5]


    plt.plot(x,y, 'bo', label='Data')
    plt.plot(x, par(x,*popt), color='r', label='Fit')
    plt.grid()
    plt.show()

    return popt, error, r_squared, V_x

def fit(x, y):
    popt, pcov = curve_fit(par,x,y)
    a,b,c = popt

    correlation_matrix = np.corrcoef(x, y)
    correlation_xy = correlation_matrix[0,1]
    r_squared = correlation_xy**2
    error = np.sqrt(np.diagonal(pcov)) #incerteza dos parametros
    x_std = np.std(x)
    V_x = -b/(2*a) # x coordinate vertex
    V_x_std = np.sqrt(((-b/(2*a))**2)*error[1]**2 + ((b/(2*a**2)**2)*error[0]**2))
    V_y = (-popt[1]**2-4*popt[0]*popt[2])/(4*popt[0]) # y coordinate vertex
    peak = (V_x,V_y)

    return popt, error, r_squared, V_x, V_x_std
```
