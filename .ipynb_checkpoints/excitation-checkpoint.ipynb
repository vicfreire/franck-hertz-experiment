**Reading the datasets**

```python
# T1 = 160 ºC: t11 for Vr = 0.5V, t12 for Vr = 1.5V and t13 for Vr = 2.5V.

t11 = np.loadtxt('Data\T160-VA35-VR05.txt')
dft11 = pd.DataFrame(t11, columns=['VA', 'Current (a.u.)', 'Temperature (ºC)'])

t12 = np.loadtxt('Data\T160-VA35-VR15.txt')
dft12 = pd.DataFrame(t12, columns=['VA', 'Current (a.u.)', 'Temperature (ºC)'])

t13 = np.loadtxt('Data\T160-VA35-VR25.txt')
dft13 = pd.DataFrame(t13, columns=['VA', 'Current (a.u.)', 'Temperature (ºC)'])

# T2 = 180 ºC: t21 for Vr = 0.5V, t22 for Vr = 1.5V and t23 for Vr = 2.5V.

t21 = np.loadtxt('Data\T180-VA35-VR05.txt')
dft21 = pd.DataFrame(t21, columns=['VA', 'Current (a.u.)', 'Temperature (ºC)'])

t22 = np.loadtxt('Data\T180-VA35-VR15.txt')
dft22 = pd.DataFrame(t22, columns=['VA', 'Current (a.u.)', 'Temperature (ºC)'])

t23 = np.loadtxt('Data\T180-VA35-VR25.txt')
dft23 = pd.DataFrame(t23, columns=['VA', 'Current (a.u.)', 'Temperature (ºC)'])

# T3 = 200 ºC: t31 for Vr = 0.5V, t32 for Vr = 1.5V and t33 for Vr = 2.5V.

t31 = np.loadtxt('Data\T200-VA35-VR05.txt')
dft31 = pd.DataFrame(t31, columns=['VA', 'Current (a.u.)', 'Temperature (ºC)'])

t32 = np.loadtxt('Data\T200-VA35-VR15.txt')
dft32 = pd.DataFrame(t32, columns=['VA', 'Current (a.u.)', 'Temperature (ºC)'])

t33 = np.loadtxt('Data\T200-VA35-VR25.txt')
dft33 = pd.DataFrame(t33, columns=['VA', 'Current (a.u.)', 'Temperature (ºC)'])
```

**Plotting**

**T = 160 ºC**

```python
plt.figure(figsize=(12,8))
plt.plot(dft11['VA'], dft11['Current (a.u.)'], color='black', lw=3, label ='Vr = 0.5 V')
plt.plot(dft12['VA'], dft12['Current (a.u.)'], color='red', lw=3, label = 'Vr = 1.5 V')
plt.plot(dft13['VA'], dft13['Current (a.u.)'], color='green', lw=3, label = 'Vr = 2.5 V')
plt.ylabel('Current (a.u.)', fontsize=15)
plt.xlabel('VA (V)', fontsize=15)
plt.grid()
plt.legend()
plt.title('T = 160 ºC');
```
